***Task 8: Investment Duration***

```python
import numpy as np
def midpoint(Duration):
    if 'More than' in Duration:
        lower_bound = int(Duration.split()[2])
        return lower_bound + 2.5
    elif 'Less than' in Duration:
        upper_bound = int(Duration.split()[2])
        return upper_bound / 2
    else:
        lower_bound, upper_bound = map(int, Duration.split()[0].split('-'))
        midpoint = (lower_bound + upper_bound) / 2
        return midpoint
df['midpoint'] = df['Duration'].apply(midpoint)
average_duration = df['midpoint'].mean()
print("Average investment duration:", average_duration,"Years")
```

***Task 10: Correlation Analysis***

```python
import pandas as pd

def mid(value):
    if 'More than' in value:
        return int(value.split()[2])+2
    elif 'Less than' in value:
        return int(value.split()[2])
    else:
        # Handle percentage values
        lower, upper = value.split('-')
        lower = int(lower.strip('%'))
        upper = int(upper.strip('%'))
        return (lower + upper) / 2

df = pd.read_csv("/content/drive/MyDrive/Details/Copy of Data_set 2 - Copy.csv")
df['Exp_mid'] = df['Expect'].apply(mid)
Exp=df['Exp_mid']
Exp
```

```python
import numpy as np
def midpoint(Duration):
    if 'More than' in Duration:
        lower_bound = int(Duration.split()[2])
        return lower_bound
    elif 'Less than' in Duration:
        upper_bound = int(Duration.split()[2])
        return upper_bound
    else:
        lower_bound, upper_bound = map(int, Duration.split()[0].split('-'))
        midpoint = (lower_bound + upper_bound) / 2
        return midpoint
Dur=df['midpoint'] = df['Duration'].apply(midpoint)
Dur
```

```python
import pandas as pd
new = {'Exp': Exp, 'Dur': Dur}
data = df.assign(**new)
data
```

```python
Pearsons_correlation = data['Exp'].corr(data['Dur'])
print(Pearsons_correlation)
```



---


As the result of the correlation between Expected returns and Duration is 0.23, it is said to be a weak correlation.
